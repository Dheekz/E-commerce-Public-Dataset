### Pertanyaan 2:

```python
#2. Berapa rata-rata nilai transaksi yang dilakukan oleh pelanggan?

order_with_payment = pd.merge(orders_df, order_payments_df, on='order_id', how='inner')

order_total_payment = order_with_payment.groupby('order_id')['payment_value'].sum()

upper_limit = np.percentile(order_total_payment, 95)

capped_order_total_payment = np.clip(order_total_payment, None, upper_limit)

average_transaction_value = capped_order_total_payment.mean()

plt.figure(figsize=(10, 6))
sns.histplot(capped_order_total_payment, kde=True)
plt.xlabel('Nilai Transaksi')
plt.ylabel('Frekuensi')
plt.title('Distribusi Nilai Transaksi (Dengan Capping)')
plt.axvline(average_transaction_value, color='red', linestyle='dashed', linewidth=2, label='Rata-rata Nilai Transaksi')
plt.legend()
plt.show()

print(f"Rata-rata Nilai Transaksi: {average_transaction_value:.2f}")
```


**Insight:**
Berdasarkan histogram yang dihasilkan, kita dapat melihat bahwa:

1. Sebagian besar transaksi memiliki nilai yang relatif rendah.
2. Distribusi nilai transaksi cenderung condong ke kanan (right-skewed), artinya ada beberapa transaksi dengan nilai yang sangat tinggi yang memengaruhi rata-rata.
3. Untuk mengurangi pengaruh outlier, capping pada nilai transaksi dilakukan dengan membatasi nilai transaksi pada persentil ke-95.
4. Rata-rata nilai transaksi setelah capping menunjukkan nilai transaksi rata-rata yang lebih representatif dan tidak dipengaruhi oleh nilai outlier.
5. Garis merah putus-putus menunjukkan rata-rata nilai transaksi yang telah dihitung setelah proses capping.

Dengan demikian, analisis ini memberikan informasi yang lebih akurat mengenai nilai transaksi rata-rata pelanggan, dengan mempertimbangkan faktor outlier yang dapat mempengaruhi hasil analisis.

## Analisis Lanjutan: RFM (Recency, Frequency, Monetary)

```python
# Analisis Lanjutan: RFM (Recency, Frequency, Monetary)

rfm_data = pd.merge(orders_df, order_payments_df, on='order_id', how='inner')
rfm_data = pd.merge(rfm_data, customers_df, on='customer_id', how='inner')

rfm_data['order_purchase_timestamp'] = pd.to_datetime(rfm_data['order_purchase_timestamp'])

last_date = rfm_data['order_purchase_timestamp'].max()

rfm_table = rfm_data.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (last_date - x.max()).days,  # Recency
    'order_id': 'count',  # Frequency
    'payment_value': 'sum'  # Monetary
})

rfm_table.rename(columns={
    'order_purchase_timestamp': 'Recency',
    'order_id': 'Frequency',
    'payment_value': 'Monetary'
}, inplace=True)

quantiles = rfm_table.quantile(q=[0.25, 0.5, 0.75])
segments = {
    'Recency': {
        'High': lambda x: x <= quantiles['Recency'][0.25],
        'Medium': lambda x: quantiles['Recency'][0.25] < x <= quantiles['Recency'][0.75],
        'Low': lambda x: x > quantiles['Recency'][0.75]
    },
    'Frequency': {
        'High': lambda x: x >= quantiles['Frequency'][0.75],
        'Medium': lambda x: quantiles['Frequency'][0.25] <= x < quantiles['Frequency'][0.75],
        'Low': lambda x: x < quantiles['Frequency'][0.25]
    },
    'Monetary': {
        'High': lambda x: x >= quantiles['Monetary'][0.75],
        'Medium': lambda x: quantiles['Monetary'][0.25] <= x < quantiles['Monetary'][0.75],
        'Low': lambda x: x < quantiles['Monetary'][0.25]
    }
}

def assign_segment(row):
    recency_segment = 'Low'
    for segment, condition in segments['Recency'].items():
        if condition(row['Recency']):
            recency_segment = segment
            break

    frequency_segment = 'Low'
    for segment, condition in segments['Frequency'].items():
        if condition(row['Frequency']):
            frequency_segment = segment
            break

    monetary_segment = 'Low'
    for segment, condition in segments['Monetary'].items():
        if condition(row['Monetary']):
            monetary_segment = segment
            break

    return recency_segment, frequency_segment, monetary_segment

rfm_table[['Recency_Segment', 'Frequency_Segment', 'Monetary_Segment']] = rfm_table.apply(assign_segment, axis=1, result_type='expand')

# Visualisasikan segmen pelanggan berdasarkan RFM
plt.figure(figsize=(12, 6))
sns.countplot(x='Recency_Segment', data=rfm_table)
plt.xlabel('Recency Segment')
plt.ylabel('Jumlah Pelanggan')
plt.title('Segmen Pelanggan Berdasarkan Recency dan Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='Frequency_Segment', data=rfm_table)
plt.xlabel('Frequency Segment')
plt.ylabel('Jumlah Pelanggan')
plt.title('Segmen Pelanggan Berdasarkan Recency dan Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='Monetary_Segment', data=rfm_table)
plt.xlabel('Monetary Segment')
plt.ylabel('Jumlah Pelanggan')
plt.title('Segmen Pelanggan Berdasarkan Monetary')
plt.show()
```


**Insight dari Analisis RFM**

- Pelanggan dengan skor Recency tinggi (High) dan Frequency tinggi (High) adalah pelanggan terbaik yang loyal dan sering berbelanja.
- Pelanggan dengan skor Recency rendah (Low) dan Frequency rendah (Low) adalah pelanggan yang sudah lama tidak berbelanja dan jarang berbelanja, perlu diberikan perhatian khusus.
- Pelanggan dengan skor Monetary tinggi (High) adalah pelanggan yang memberikan pendapatan terbesar, perlu diberikan program loyalitas dan retensi."
- Pelanggan dengan skor Monetary rendah (Low) adalah pelanggan yang memberikan pendapatan kecil, perlu ditingkatkan dengan promosi atau penawaran khusus."

Namun, analisis lanjutan diperlukan untuk menyempurnakan strategi bisnis, terutama dengan mempertimbangkan faktor eksternal seperti kompetisi dan dinamika pasar.
